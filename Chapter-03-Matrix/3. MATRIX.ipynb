#  3.9n Unitary matrix example (complex numbers)

```python
# Unitary matrix example (complex numbers)
F = np.array([[1/np.sqrt(2), 1j/np.sqrt(2)], [-1j/np.sqrt(2), 1/np.sqrt(2)]])
unitary_check = np.allclose(np.dot(np.conjugate(F.T), F), np.eye(2))
print("Unitary Matrix:\n", F)
print("Is Unitary:", unitary_check)
```

# 3.9o Adjoint of a Matrix

```python
# Example matrix
matrix = np.array([[1, 2], 
                   [3, 4]])

# Function to calculate the cofactor matrix
def cofactor_matrix(matrix):
    # Calculate the determinant of the matrix
    det = np.linalg.det(matrix)
    
    # If determinant is zero, adjoint does not exist
    if det == 0:
        return None
    
    # Calculate the cofactor matrix (using minor matrix and determinant)
    cofactors = []
    for i in range(matrix.shape[0]):
        row = []
        for j in range(matrix.shape[1]):
            # Minor matrix: remove row i and column j
            minor = np.delete(np.delete(matrix, i, axis=0), j, axis=1)
            cofactor = (-1)**(i+j) * np.linalg.det(minor)
            row.append(cofactor)
        cofactors.append(row)
    
    # Return the cofactor matrix
    return np.array(cofactors)

# Compute the cofactor matrix
cofactors = cofactor_matrix(matrix)

# If cofactor matrix exists, compute the adjoint (transpose of cofactor matrix)
if cofactors is not None:
    adjoint = cofactors.T
    print("Adjoint of the Matrix:\n", adjoint)
else:
    print("Adjoint does not exist (determinant is 0)")
```

# 3.9u Minor of an Element a_{ij}​  in ∣𝐴∣

```python
def minor(matrix, i, j):
    # Delete the i-th row and j-th column
    submatrix = np.delete(np.delete(matrix, i, axis=0), j, axis=1)
    return np.linalg.det(submatrix)

# Calculate the minor of the element at position (0, 0) in A
minor_00 = minor(A, 0, 0)
print("Minor of element a[0][0]:", minor_00)
```

# 3.9v Cofactor of an Element a_{ij}​ in ∣𝐴∣

```python
def cofactor(matrix, i, j):
    # Calculate the minor
    minor_val = minor(matrix, i, j)
    # Apply the sign factor (-1)^(i+j)
    return (-1) ** (i + j) * minor_val

# Calculate the cofactor of the element at position (0, 0) in A
cofactor_00 = cofactor(A, 0, 0)
print("Cofactor of element a[0][0]:", cofactor_00)
```
