# Joint, Marginal and Conditional Probability Distributions

**3.6d Joint Probability Density Function (PDF)**

```python
from scipy.stats import multivariate_normal

# Assuming height and canopy_vol follow a bivariate normal distribution
mean_vector = df[['height', 'canopy_vol']].mean().values
cov_matrix = df[['height', 'canopy_vol']].cov().values

# Calculate the PDF for each observation
pdf_values = multivariate_normal.pdf(df[['height', 'canopy_vol']], mean=mean_vector, cov=cov_matrix)
df['pdf'] = pdf_values

print("\nJoint Probability Density Function (PDF) Values:")
print(df[['height', 'canopy_vol', 'pdf']])
```

**3.6e Marginal Probability**

```python
# Marginal probability for height (integrate over canopy_vol)
marginal_height = df.groupby('height')['pdf'].sum()
print("\nMarginal Probability for Height:")
print(marginal_height)

# Marginal probability for canopy_vol (integrate over height)
marginal_canopy = df.groupby('canopy_vol')['pdf'].sum()
print("\nMarginal Probability for Canopy Volume:")
print(marginal_canopy)
```

1.   Height 16.172: The marginal probability of observing a height of 16.172 is approximately 0.393. This means that when you consider all possible values of canopy_vol, the likelihood of encountering a height of 16.172 is about 39.3%.
2.   Height 18.966: The marginal probability of observing a height of 18.966 is very low (0.015). This indicates that such a height is quite rare in your dataset, with only about a 1.5% chance of occurrence.

**3.6f Conditional Probability**

```python
# Conditional probability P(height | canopy_vol)
# Assuming height is the dependent variable

df['conditional_prob'] = df['pdf'] / df['canopy_vol'].map(marginal_canopy)

print("\nConditional Probability P(height | canopy_vol):")
print(df[['height', 'canopy_vol', 'conditional_prob']])
```

# Wishart Distribution

```python
# Compute the sample covariance matrix from df2
cov_matrix_mle = np.cov(data, rowvar=False)

# Number of variables (dimensions)
p = cov_matrix_mle.shape[0]

# Degrees of freedom for Wishart distribution (should be >= p)
df = p + 2  # Example value; typically df > p

# Define the Wishart distribution with the sample covariance matrix
wishart_dist = wishart(df=df, scale=cov_matrix_mle)

# Generate a sample from the Wishart distribution
wishart_sample = wishart_dist.rvs(size=1)

print("Sample from the Wishart Distribution:\n", wishart_sample)

# Compute the mean and covariance matrix of the Wishart sample (for verification)
mean_wishart = np.mean(wishart_sample, axis=0)
cov_wishart = np.cov(wishart_sample[0], rowvar=False)

print("Mean of Wishart Sample:\n", mean_wishart)
print("\nCovariance Matrix of Wishart Sample:\n", cov_wishart)
```
